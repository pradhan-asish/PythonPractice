```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
html_data = requests.get('https://en.wikipedia.org/wiki/List_of_largest_banks')
```

```python
html_data.text[101:124]
```

```python
soup = BeautifulSoup(html_data.content)
```

```python
data = pd.DataFrame(columns=["Name","Market Cap (US$ Billion)"])
```

```python
for row in soup.find_all('tbody')[3].find_all('tr'):
    col = row.find_all('td')
    for col1 in col:
        for col2 in col1:
            print(col2)
```

```python
index = []
bank_name = []
value = []
for row in soup.find_all('tbody')[3].find_all('tr'):
    col = row.find_all('td')
    if (len(col) != 0) :
        i = 0
        for col1 in col:
            if (i==0) :
                index.append(col1.get_text())
            elif(i==1) :
                bank_name.append(col1.get_text(). rstrip("\n"))
            elif(i==2) :
                value.append(col1.get_text(). rstrip("\n"))
            i += 1

d = {'Name': bank_name, 'Market Cap (US$ Billion)': value}
print(d)

data = pd.DataFrame( data = d)
```
